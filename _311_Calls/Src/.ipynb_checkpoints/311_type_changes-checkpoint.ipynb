```python
def get311data(key, start_time, end_time):
    client = Socrata("data.cityofnewyork.us", key)
    results = client.get("fhrw-4uyv", where = "created_date BETWEEN " + start_time + " AND " +  end_time, limit = 1000000000)
    
    return(pd.DataFrame.from_records(results))
```

# Quantiy Change

```python
quantity_before = brooklyn_before.groupby('complaint_type').count()[['borough']].reset_index()
quantity_after = brooklyn_after.groupby('complaint_type').count()[['borough']].reset_index()
```

```python
qttcBrooklyn = quantity_before.merge(quantity_after, on = "complaint_type").rename(columns = {'borough_x': 'before', 'borough_y': 'after'})
```

```python
qttcBrooklyn['qchange']  = qttcBrooklyn['after'] - qttcBrooklyn['before'] 
```

```python
qttcBrooklyn = qttcBrooklyn.sort_values(by = ['qchange'], ascending = False)
```

```python
qttcBrooklyn.head()
```

```python
quantity_before = manhattan_before.groupby('complaint_type').count()[['borough']].reset_index()
quantity_after = manhattan_after.groupby('complaint_type').count()[['borough']].reset_index()

qttcManhattan = quantity_before.merge(quantity_after, on = "complaint_type").rename(columns = {'borough_x': 'before', 'borough_y': 'after'})
qttcManhattan['qchange']  = qttcManhattan['after'] - qttcManhattan['before'] 
qttcManhattan = qttcManhattan.sort_values(by = ['qchange'], ascending = False)
```

```python
qttcManhattan.head()
```

```python
quantity_before = bronx_before.groupby('complaint_type').count()[['borough']].reset_index()
quantity_after = bronx_after.groupby('complaint_type').count()[['borough']].reset_index()

qttcBronx = quantity_before.merge(quantity_after, on = "complaint_type").rename(columns = {'borough_x': 'before', 'borough_y': 'after'})
qttcBronx['qchange']  = qttcBronx['after'] - qttcBronx['before'] 
qttcBronx = qttcBronx.sort_values(by = ['qchange'], ascending = False)
```
